# Merge two different data into single Dataframe

```python
final_df=pd.merge(df,df_country,on="Country Code",how="left")
```

```python
final_df.head()
```

```python
Country_Name=final_df["Country"].value_counts().index
```

```python
Country_val=final_df["Country"].value_counts().values
```

# Top 3 Countries that uses Zomato

```python
plt.pie(Country_val[:3],labels=Country_Name[:3],autopct="%1.2f%%")
plt.show()
```

Observation:Maximum Transactions of Zomato are from India followed by USA and UK

```python
final_df.columns
```

# Ratings

```python
ratings=final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
ratings
```

Observation:
1.When the Rating is between 4.5 to 4.9 --> Excellent
2.When the Rating is between 4.0 to 4.4 --> Very Good
3.When the Rating is between 3.5 to 3.9 --> Good
4.When the Rating is between 2.5 to 3.4 --> Average
5.When the Rating is between 1.8 to 2.4 --> Poor

```python
fig=plt.figure(figsize=(12,6))
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings,hue="Rating color",palette=["black","red","Orange","yellow","green","green"])
plt.title("Aggregate rating vs Rating Count")
```

Observation:Maximum number of Rating are between 2.5 and 3.4

```python
sns.countplot(x="Rating color",data=ratings,palette=["black","red","Orange","yellow","green","green"])
```
